#### Here we will demonstrate the memory structure for experience replay.<br>
The deque function creates a First in First out (FIFO) memory structure, here we choose an arbitrary length of 7 (when we do this for real we will use a much larger memory buffer). Each time an item is appended to the queue the others get pushed toward the exit point. If the queue is full and another item is appended the first item gets pushed out. 

```python
memory = deque([], maxlen=7)
```

We can now use our model to generate action value predictions for the obervations at time t; this is the equivalent to the Q(St,at) in the Q table. <br>
And action value predictions for the obervations at time t + 1; this will be used in the Bellman optimality equation just like in the Q table but here we use it to create the label...

```python
prediction_at_t = model.predict(obs_t)
prediction_at_t
```

```python
prediction_at_t_next = model.predict(obs_t_next)
prediction_at_t_next
```

Now that we have action values for time t and t+1 we can create training data for our ANN.<br>
Each observation at time t becomes our feature data X (4 features representing the vector of observations).
The y_hat data is the action value vector (note: there are only 2 actions in this environment).
The prediction at time t+1 is used in the Bellman optimality equation: <br><br>
<center><b> target = reward + gamma * np.max(prediction_at_t_next[i])</b></center><br>
The target value calculated is used to update the action value Q(St,at), it is these updated action values that are used as our label in the training data.<br><br> 
<b>Note: The target is just the reward if the episode is done.</b><br><br>
The actual update, as it relates to the Q table, is carried out by the ANN (the weights) as it takes the mean squared difference between the predicted action values from observations at time t and the updated action values (y).
If you don't understand this please watch the videos for this week.<br> The following cell shows the creation of labelled training data, make sure you understand this and how it relates to the videos...

```python
X = []          #This is our feature vector, the most recent observation
y = []          #This is our label calculated using the long term discounted reward
gamma = 0.9     #This is our discount factor
i = 0
for obs_t, action, reward, _, done in batch_vector:#get a row from our batch
    
    X.append(obs_t)                                #append the most recent observation to X 
    
    if done:                                       #if the episode was over
        target = reward                            #the target value is just the reward
    else:                                          #otherwise
        #the target value is the discounted optimal reward (Bellman optimality equation)
        #Remember we use the max action_value from the state(observation) at time t+1
        target = reward + gamma * np.max(prediction_at_t_next[i])
    
    #now we update the action value for the original state(observation) given the action that
    #was taken, and we use the target value as the update.
    prediction_at_t[i,action] = target
    #the updated action values are used as the label 
    y.append(prediction_at_t[i])        #Remember the update will be used as the label to update the ANN weights
                                        #by backpropagating the mean squared error. 
    i += 1
```

After this our feature data for our batch is...

```python
X = np.array(X).reshape(batch_size, obs_count)
X
```

We'll need to convert our labels into a numpy array

```python
y = np.array(y)
y
```

<b>Exercise 2: Experience Replay Function</b>

You now have enough information to create an experience replay function: <br><br>
<center><b>experience_replay(model, batch_size, gamma, memory, obs_count, action_count, epoch_count, loss)</b></center><br><br>
The function should take in a model, batch_size, a memory structure, the number of observations from the environment, the number of actions from the environment, the number of epochs to train the ANN on, and a list of losses (the list of losses will be appended to each time the action replay function runs). The function should return the updated loss list.

```python
def experience_replay(model, batch_size, gamma, memory, obs_count, action_count, epoch_count, loss):
    batch = random.sample(memory, batch_size) #sample a batch from the experience memory using batch_size
    batch_vector = np.array(batch, dtype=object) #vectorise the batch
    
    obs_t = np.zeros(shape = (batch_size, obs_count)) #observation at t: create a numpy array of zeros with dimensions batch_size, state_count
    obs_t_next = np.zeros(shape = (batch_size, obs_count)) #observation at t+1: create a numpy array of zeros with dimensions batch_size, state_count
    
    for i in range(len(batch_vector)): #loop through the batch collecting the obs at t and obs at t+1
        obs_t[i] = batch_vector[i,0] #store the observations at t in the relevant array
        obs_t_next[i] = batch_vector[i,3] #store the observations at t+1 in the relevant array
    
    prediction_at_t = model.predict(obs_t) #Use the model to predict an action using observations at t
    prediction_at_t_next = model.predict(obs_t_next) #Use the model to predict an action using observations at t+1
    
    #Create the features(X) and lables(y)
    X = [] #This is our feature vector, the most recent observation                  
    y = [] #This is our label calculated using the long term discounted reward
    
    i = 0
    
    for obs_t, action, reward, _, done in batch_vector: #get a row from our batch
        
        X.append(obs_t) #append the most recent observation to X 
        
        if done:                                 #if the episode was over
            target = reward #the target value is just the reward
        else:                                    #otherwise
            #the target value is the discounted optimal reward (Bellman optimality equation)
            #Remember we use the max action_value from the state(observation) at time t+1
            target = reward + gamma * np.max(prediction_at_t_next[i]) #<- One line of code here

        #now we update the action value for the original state(observation) given the action that
        #was taken, and we use the target value as the update.
        prediction_at_t[i,action] = target #<- One line of code here
        #the updated action values are used as the label 
        y.append(prediction_at_t[i]) #<- One line of code here, remember the update will be used as the label to update 
                                        #the ANN weights by backpropagating the mean squared error. 

        i += 1 #increment i

    X = np.array(X).reshape(batch_size, obs_count) #reshape X
    y = np.array(y) #create a numpy array from y
    hist = model.fit(X, y, epochs = epoch_count) #fit the model with X,y and epochs
    loss = []
    for i in range(epoch_count): #loop over the epoch_count
        loss.append(hist.history['loss'][i]) #record the loss for analysis 
        
    return loss #return the loss
    
```

Now we have our experience replay function to organise our RL problem into a supervised learning problem, we can use a larger memory buffer, fill this up with real environment experience and use a large batch size to sample the experience, and train our DQN.<br>

The code below demostrates the epsilon-greedy policy, you should recognise the parts of code that <b>explore/exploit</b> by now. The agent builds up experience by both randomly exploring as well as exploiting the environment. Once there is enough (batch_size) experience in memory the DQN starts training on the data and continues every environment step on a batch of experience that is randomly sampled from memory... <br>
<br>Note: Watch the output i.e. the first episodes will be short as the agent explores and is untrained when exploiting, resulting in early failures (short eposides). The episodes should get longer over time though, meaning the agent is getting better at balancing the pole and is exploiting more. 

<b>Note:</b> I have set the episodes to 50 i.e. sometimes the agent can get good enough to achieve awards up to 200 (200 steps) within 50 but sometimes not, so just run the cell again i.e. the previous weights will be trained for a further 50 and so on...<br>

<b>Note 2:</b> When the agent gets good the episodes can seem long and boring! when you are getting rewards above 150 you can interupt the kernel in the notebook (Kernel -> Interrupt) and your model should maintain it's weight values.


Let's recreate our model first so we can see everything in the same place...

```python
alpha = 0.001
model = keras.Sequential()
model.add(keras.layers.Dense(24, input_dim=obs_count, activation='relu'))
model.add(keras.layers.Dense(24, activation='relu'))
model.add(keras.layers.Dense(action_count, activation = 'linear'))
model.compile(loss = 'mean_squared_error', optimizer=keras.optimizers.Adam(learning_rate=alpha))
```

```python
rewards = []                                              #store the rewards in a list for analysis only
loss = []                                                 #store the losses in a list for analysis only
epsilons = []
episodes = 50                                             #We will start with 50 episodes and see how it goes
gamma = 0.9                                               #This is the discount rate
beta = 0.3                                                #This is the epsilon decay rate
batch_size = 16                                           #The batch size for solving the IID problem
memory = deque([], maxlen=2500)                           #The memory replay buffer set to a length of2500 

#Please read through this code and try and understand what is happening, you should understand this 
#if you have watched the videos but if you need any help ask the lab supervisor.
for episode in range(episodes):
    obs_t = cartpole_env.reset()
    obs_t = np.reshape(obs_t[0], [1, obs_count])
    
    total_reward = 0
    epsilon = 1 / (1 + beta * (episode / action_count))
    
    done = False
    while not done:
        rand_num = np.random.random()
        if rand_num <= epsilon:
            action = cartpole_env.action_space.sample()
        else:
            action_values = model.predict(obs_t, verbose=0)
            action = np.argmax(action_values[0])
    
        obs_t_next, reward, done,_, info = cartpole_env.step(action)
        obs_t_next = np.reshape(obs_t_next, [1, obs_count])
        total_reward += reward
        memory.append((obs_t, action, reward, obs_t_next, done))
        obs_t = obs_t_next
    
        if done:
            rewards.append(total_reward)
            epsilons.append(epsilon)
            print(f'episode: {episode}/{episodes}, score: {total_reward}, epsilon: {epsilon}')
    
        if len(memory) > batch_size:
            loss = experience_replay(model, batch_size, gamma, memory, obs_count, action_count, 1, loss)
        
    rewards.append(total_reward)   
```

Let's view the loss in a graph:

```python
plt.figure(figsize=(15, 5))
plt.title('Loss Graph')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.plot(range(len(loss)),loss)
plt.show()
```

This does show that the losses are reducing over time but because we are using an epsilon greedy policy ther will be spikes where we are exporing. Let's view the rewards in a graph, this may be less messy...

```python
plt.figure(figsize=(15, 5))
plt.title('Rewards Graph')
plt.xlabel('Episodes')
plt.ylabel('Rewards')
plt.plot(range(len(rewards)),rewards)
plt.show()
```

We can clearly see that the agent is learning how to maximise it's reward.
